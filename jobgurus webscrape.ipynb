```python
#import your libraries
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
from time import sleep
from random import randint
import numpy as np
```

```python
URL = "https://www.jobgurus.com.ng/jobs/index/2"
#requesting the URL above
page = requests.get(URL)

#specifying format of the page to help readability
soup= BeautifulSoup(page.text, "html.parser")

#printing some characters from soup in a more structured format
print(soup.prettify()[:200])
```

```python
#creating a function to extract the title
def extract_title(soup):
    title = []
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for h2 in div.find_all(name='h2'):
            for a in h2.find_all(name='a', attrs={'href':True}):
                title.append(a.text.strip())
    return(title)
extract_title(soup)
```

```python
#creating function to extract specifications
def extract_spec(soup):
    spec = []
    
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for s in div.find_all(name='strong'):
            for p in s.find_all(name='p', attrs={'class':'job-info'}):
                for a in s.find_all(name='a', attrs={'href':True}):
                    spec.append(a.text.strip())
    return(spec)
extract_spec(soup)
```

```python
#creating function to extract year
def extract_year(soup):
    year = []
    for div in soup.find_all(name='div', attrs={'class':'panel-footer clearfix'}):
        for div in div.find_all(name='div', attrs={'class':'job-footer_text'}):
            for span in div.find_all(name='span'):
                year.append(span.text.strip())
            
    return(year)
extract_year(soup)
```

```python
#creating function for extracting job brief
def extract_brief(soup):
    brief = []
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for div in div.find_all(name='div', attrs={'class':'job-brief'}):
            brief.append(div.text.strip())
            
    return(brief)
extract_brief(soup)
```

```python
#scraping code
#creating dict to hold values
data = {'Job_title': [], 'Spec': [], 'Date_posted': [], 'Job_brief': []}

#specifing the pages to scrape from
pages = np.arange(2, 90, 1)

for page in pages:
    page = requests.get("https://www.jobgurus.com.ng/jobs/index/" + str(page))
    
    soup = BeautifulSoup(page.text, 'html.parser')
    sleep(randint(2,10))#ensuring at least 2 seconds between page grabs
    
    #grabbing job titles
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for h2 in div.find_all(name='h2'):
            for a in h2.find_all(name='a', attrs={'href':True}):
                data['Job_title'].append(a.text.strip())
    
    #grabbing specs
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for s in div.find_all(name='strong'):
            for p in s.find_all(name='p', attrs={'class':'job-info'}):
                for a in s.find_all(name='a', attrs={'href':True}):
                    data['Spec'].append(a.text
                                        .strip())
     
    #grabbing dates
    for div in soup.find_all(name='div', attrs={'class':'panel-footer clearfix'}):
        for div in div.find_all(name='div', attrs={'class':'job-footer_text'}):
            for span in div.find_all(name='span'):
                data['Date_posted'].append(span.text.strip())
                
    #grabbing job brief
    for div in soup.find_all(name='div', attrs={'class':'panel-body'}):
        for div in div.find_all(name='div', attrs={'class':'job-brief'}):
            data['Job_brief'].append(div.text.strip())
            
print('Done')
```

```python
#creating new list to separate company,location and specialization data which were in values in the same dict
N = data['Spec']
comp_list = N[::3]
spec_list = N[1::3]
loc_list = N[2::3]
```

```python
#deleting key holding specs
data1 = data.pop('Spec')
```

```python
#creating dataframe from the dict
df1 = pd.DataFrame(data, columns=['Job_title', 'Date_posted', 'Job_brief'])
df1.head()
```

```python
#creating dataframe from lists created
df2 = pd.DataFrame(list(zip(spec_list, comp_list, loc_list)), columns = ['Specialization', 'Company', 'Location'])
df2.head()
```

```python
#concating dataframes
jobs_df = pd.concat([df1, df2], axis=1, sort=False)
jobs_df.head()
```

```python
#splitting job title column to get job title without company
df_21 = jobs_df.Job_title.str.split(' Job ',1, expand=True).rename({0:'Job', 1:'Re.Company'}, axis=1).to_dict()
#creating a dataframe to hold values
df_job1 = pd.DataFrame(data=df_21)
df_job1.head()
```

```python
#concat new dataframes
df_jobs = pd.concat([df_job1, jobs_df], axis=1, sort=False)
df_jobs.head()
```
